# 作業1

```python
from pymongo import MongoClient
```

```python
client = MongoClient(host='127.0.0.1', port=27017)
```

```python
db = client['AIoT_Day08']
```

```python
collection = db['member']
```

# 作業2
* 使用 Python 操作 Mongo 中的資料。

```python
mydata = {'name':'A', 'phone':'0123456789', 'email':'abc@gmail.com'}
```

```python
result = collection.insert_one(mydata)
```

```python
print(result.inserted_id)
```

```python
data_list = [
    {'name':'B', 'phone':'9876543210', 'email':'def@gmail.com'},
    {'name':'C', 'phone':'1212121212', 'email':'ghi@gmail.com'}
]
```

```python
result = collection.insert_many(data_list)
```

```python
result.inserted_ids
```

```python
filter_ = {'name':'C'}
update_ = {"$set":{'name':'D'}}
collection.update_one(filter_, update_)
```

```python
collection.delete_one({'name':'B'})
```
